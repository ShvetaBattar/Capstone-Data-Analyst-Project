### Task 1: Visualizing Data Composition with Pie Charts


##### 1.1 Create a Pie Chart of the Top 5 Databases Respondents Want to Work With


In the survey data, the `DatabaseWantToWorkWith` column lists the databases that respondents wish to work with. Let’s visualize the top 5 most-desired databases in a pie chart.



```python
##Write your code here

# Count the occurrences of each database in the 'DatabaseWantToWorkWith' column
database_counts = df['DatabaseWantToWorkWith'].value_counts()

# Select the top 5 most desired databases
top_5_databases = database_counts.head(5)

# Plot the Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_databases, labels=top_5_databases.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Databases Respondents Want to Work With', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.

# Show the plot
plt.show()
```

The `DevType` column lists the developer types for respondents. We’ll examine the distribution by showing the top 5 developer roles in a pie chart.



```python
##Write your code here


# Count the occurrences of each developer role in the 'DevType' column
devtype_counts = df['DevType'].value_counts()

# Select the top 5 most common developer roles
top_5_devtypes = devtype_counts.head(5)

# Plot the Pie Chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_devtypes, labels=top_5_devtypes.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top 5 Developer Roles', fontsize=16)
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.

# Show the plot
plt.show()
```

##### 1.3 Create a pie chart for the operating systems used by respondents for professional use


The `OpSysProfessional` use column shows the operating systems developers use professionally. Let’s visualize the distribution of the top operating systems in a pie chart.



```python
##Write your code here

# Step 1: Clean the 'OpSysProfessional use' column by handling missing values (NA)
df['OpSysProfessional use'] = df['OpSysProfessional use'].fillna('')  # Replace NA with empty string

# Step 2: Split the column by ';' to separate multiple OS entries
os_list = df['OpSysProfessional use'].str.split(';', expand=True).stack()

# Step 3: Count the frequency of each operating system
os_counts = os_list.value_counts()

# Step 4: Plot the top 5 operating systems in a pie chart
top_5_os = os_counts.head(5)

plt.figure(figsize=(8, 8))
top_5_os.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0'])

# Add title and adjust layout
plt.title('Top 5 Operating Systems Used Professionally', fontsize=14)
plt.ylabel('')  # Remove y-label for clarity
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 2: Additional Visualizations and Comparisons


##### 2.1 Pie Chart for Top 5 Programming Languages Respondents Have Worked With


The `LanguageHaveWorkedWith` column contains the programming languages that respondents have experience with. We’ll plot a pie chart to display the composition of the top 5 languages.



```python
##Write your code here
# Step 1: Clean the 'LanguageHaveWorkedWith' column
df['LanguageHaveWorkedWith'] = df['LanguageHaveWorkedWith'].fillna('')  # Handle missing values (NA)

# Step 2: Split the column by ';' to separate multiple languages
languages = df['LanguageHaveWorkedWith'].str.split(';', expand=True).stack()

# Step 3: Count the frequency of each programming language
language_counts = languages.value_counts()

# Step 4: Get the top 5 most common programming languages
top_5_languages = language_counts.head(5)

# Step 5: Plot the pie chart
plt.figure(figsize=(8, 8))
top_5_languages.plot(kind='pie', autopct='%1.1f%%', startangle=90, 
                     colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'])

# Add title and adjust layout
plt.title('Top 5 Programming Languages Respondents Have Worked With', fontsize=14)
plt.ylabel('')  # Remove y-label for better clarity
plt.tight_layout()

# Show the plot
plt.show()
```

##### 2.2 Pie Chart for Top Collaboration Tools used in Professional Use


Using the `NEWCollabToolsHaveWorkedWith` column, we’ll identify and visualize the top collaboration tools respondents use in their professional work.



```python
##Write your code here
# Step 1: Clean the 'NEWCollabToolsHaveWorkedWith' column
df['NEWCollabToolsHaveWorkedWith'] = df['NEWCollabToolsHaveWorkedWith'].fillna('')  # Handle missing values (NA)

# Step 2: Split the column by ';' to separate multiple collaboration tools
collab_tools = df['NEWCollabToolsHaveWorkedWith'].str.split(';', expand=True).stack()

# Step 3: Count the frequency of each collaboration tool
collab_tools_counts = collab_tools.value_counts()

# Step 4: Get the top 5 most common collaboration tools
top_5_collab_tools = collab_tools_counts.head(5)

# Step 5: Plot the pie chart
plt.figure(figsize=(8, 8))
top_5_collab_tools.plot(kind='pie', autopct='%1.1f%%', startangle=90, 
                         colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'])

# Add title and adjust layout
plt.title('Top 5 Collaboration Tools Used in Professional Work', fontsize=14)
plt.ylabel('')  # Remove y-label for better clarity
plt.tight_layout()

# Show the plot
plt.show()
```

### Task 3: Analyzing and Interpreting Composition


In this task, you will create additional pie charts to analyze specific aspects of the survey data. Use `pandas` and `matplotlib` to complete each task and interpret the findings.



##### 3.1 Pie Chart of `Respondents` Most Admired Programming Languages


The `LanguageAdmired` column lists the programming languages respondents admire most. Create a pie chart to visualize the top 5 admired languages.



```python
##Write your code here
# Step 1: Clean the 'LanguageAdmired' column
df['LanguageAdmired'] = df['LanguageAdmired'].fillna('')  # Handle missing values (NA)

# Step 2: Split the column by ';' to separate multiple admired languages
admired_languages = df['LanguageAdmired'].str.split(';', expand=True).stack()

# Step 3: Count the frequency of each admired language
admired_languages_counts = admired_languages.value_counts()

# Step 4: Get the top 5 most admired languages
top_5_admired_languages = admired_languages_counts.head(5)

# Step 5: Plot the pie chart
plt.figure(figsize=(8, 8))
top_5_admired_languages.plot(kind='pie', autopct='%1.1f%%', startangle=90, 
                              colors=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c2c2f0'])

# Add title and adjust layout
plt.title('Top 5 Admired Programming Languages', fontsize=14)
plt.ylabel('')  # Remove y-label for better clarity
plt.tight_layout()

# Show the plot
plt.show()
```

##### 3.2 Pie Chart of Tools Used for AI Development


Using the `AIToolCurrently` Using column, create a pie chart to visualize the top 5 tools developers are currently using for AI development.



```python
##Write your code here
# Step 1: Clean and preprocess the data
df['AIToolCurrently Using'] = df['AIToolCurrently Using'].fillna('')  # Handle missing (NA) values

# Split the column by ';' and flatten the resulting lists into a single series
tools_series = df['AIToolCurrently Using'].str.split(';').explode().str.strip()

# Step 2: Count the frequency of each tool
tool_counts = tools_series.value_counts()

# Step 3: Get the top 5 tools
top_5_tools = tool_counts.head(5)

# Step 4: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_tools, labels=top_5_tools.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Top 5 Tools Used for AI Development')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```

##### 3.3 Pie Chart for Preferred Web Frameworks


The `WebframeWantToWorkWith` column includes web frameworks that respondents are interested in working with. Visualize the top 5 frameworks in a pie chart.



```python
##Write your code here
# Step 1: Clean and preprocess the data
df['WebframeWantToWorkWith'] = df['WebframeWantToWorkWith'].fillna('')  # Handle missing (NA) values

# Split the column by ';' and flatten the resulting lists into a single series
frameworks_series = df['WebframeWantToWorkWith'].str.split(';').explode().str.strip()

# Step 2: Count the frequency of each framework
framework_counts = frameworks_series.value_counts()

# Step 3: Get the top 5 frameworks
top_5_frameworks = framework_counts.head(5)

# Step 4: Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_5_frameworks, labels=top_5_frameworks.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Top 5 Preferred Web Frameworks')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is circular.
plt.show()
```

##### 3.4 Pie Chart for Most Desired Embedded Technologies


Using the `EmbeddedWantToWorkWith` column, create a pie chart to show the top 5 most desired embedded technologies that respondents wish to work with.



```python
##Write your code here
# Step 1: Clean the 'EmbeddedWantToWorkWith' column
df['EmbeddedWantToWorkWith'] = df['EmbeddedWantToWorkWith'].fillna('')  # Handle missing values

# Step 2: Split multiple technologies into individual entries
technologies_series = df['EmbeddedWantToWorkWith'].str.split(';', expand=True).stack()

# Step 3: Count the occurrences of each embedded technology
technology_counts = technologies_series.value_counts()

# Step 4: Select the top 5 most frequent technologies
top_5_technologies = technology_counts.head(5)

# Step 5: Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(top_5_technologies, labels=top_5_technologies.index, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Top 5 Most Desired Embedded Technologies')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
